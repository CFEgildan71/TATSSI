```python
import os
import xarray as xr
import numpy as np
import gdal

import matplotlib
matplotlib.use('nbagg')
import matplotlib.pyplot as plt
import matplotlib.colors as colors
```

```python
DATADIR = "/home/glopez/Projects/TATSSI/data/MOD13A2.006"
fname = "MOD13A2.A2018145.h09v07.006.2018162000027.VI_QA_Land_Water_Mask.tif"
fname = os.path.join(DATADIR, fname)

# Open dataset
qa = xr.open_rasterio(fname)
```

```python
# Get categories
ds = gdal.Open(fname)
band = ds.GetRasterBand(1)
band.GetRasterColorTable()
category_names = np.array(band.GetRasterCategoryNames())
idx = np.where(category_names != '')
```

```python
qa[0].plot.pcolormesh(levels=idx[0])
plt.show()
```
