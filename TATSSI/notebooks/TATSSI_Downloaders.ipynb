## Download data
TATSSI allows to dowload data products from the Land Processes Distributed Active Archive Center [LP DAAC](https://lpdaac.usgs.gov/) derived from different sensors such as:

* Moderate Resolution Imaging Spectroradiometer [MODIS](https://lpdaac.usgs.gov/data/get-started-data/collection-overview/#MODIS_anchor)
* Visible Infrared Imaging Radiometer Suite [VIIRS](https://lpdaac.usgs.gov/data/get-started-data/collection-overview/#S-NPPVIIRS_anchor)

Downloading products from the LP DAAC requires a NASA [EarthData](https://urs.earthdata.nasa.gov/) login, please, first regsiter as a user to get the login credentials. Edit the ```config.json``` file with those details.

1. Select your are of study, you have to select a valid tile from the standard MODIS/VIIRS sinusoidal grid. You can have an overview in [here](https://earth.app.goo.gl/KBwn3C).
2. Click on the ```Select product and dates``` button. Explore the available products, double click on the row of the one you want to download
3. Select a date range
4. Once you selected a product and dates to download, set the output directory. Click on the ```Select output directory``` to do it. You can create a new directory if needed.
5. Get your data! Just click on the ```Download product for selected dates```
