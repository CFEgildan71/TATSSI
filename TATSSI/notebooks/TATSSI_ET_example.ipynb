```python
# Import TATSSI libraries
from helpers import utils
from TATSSI.notebooks.helpers.time_series_smoothing import TimeSeriesSmoothing
```

```python
# Load PET data
fname = '/home/glopez/Projects/TATSSI/data/MYD16A2.006/PET_500m/interpolated/MYD16A2.006._PET_500m.smooth.tif'
tss = None
tss = TimeSeriesSmoothing(fname=fname)
```

```python
# PET
tss.ts.data._PET_500m
```

```python
import matplotlib.pyplot as plt

# Mask non data vals
PET = tss.ts.data._PET_500m
annual_mean_PET = PET.where(PET < 32760)

# Annual composite 
annual_mean_PET = annual_mean_PET.resample(time='1Y').mean('time')
annual_mean_PET[0].plot.imshow(vmin=0.0, vmax=700.0, cmap='Blues')
plt.show()
```
